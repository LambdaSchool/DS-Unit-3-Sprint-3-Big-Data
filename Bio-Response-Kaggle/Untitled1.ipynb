```python
from load_data import X,y, X_train, X_test, y_train, y_test
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import RobustScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.feature_selection import f_regression, SelectKBest
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd

X.head()

#print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
#X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.14)


B = 6
n = int(np.divide(X.shape[0]-1, B))

# Define an estimator and param_grid
pipe = Pipeline(
    ('scale', RobustScaler()), 
    ('reduce_dim', SelectKBest(f_regression)), 
    ('classifier', SGDClassifier(X_train, y_train, verbose=10)))

param_grid = {
    'reduce_dim__k': range(B, n), 
    'classifier__loss': ['log', 'modified_huber']
}

# Fit on the train set, with grid search cross-validation
gs = GridSearchCV(pipe, param_grid=param_grid, cv=3, 
                  scoring='neg_log_loss', 
                  verbose=1, n_jobs=2)

gs.fit(X_train, y_train)
validation_score = gs.best_score_
print()
print('Cross-Validation Score:', -validation_score)
print()
print('Best estimator:', gs.best_estimator_)
print()
```
