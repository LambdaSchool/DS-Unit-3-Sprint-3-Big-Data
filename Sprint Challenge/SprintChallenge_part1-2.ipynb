# Part 2 - Big Data Options

During my time as a student at Lambda School, I have learned a great deal about a plethora of tools made for working with 'Big Data.' Before I discuss any of the actual tools, I'd like to discuss the phrase 'Big Data.' Many would consider this phrase a buzzword, highly overused. I would consider 'Big Data' any dataset that doesn't comfortably fit in memory of a single computer. Tools like AWS SageMaker are great for situations where a dataset doesn't comfortably fit in memory on a local computer. AWS SageMaker provides a way to greatly increase memory and computing power, while remaining on a single computer. There are multiple benefits of working with data on a single computer. Firstly, trying to distribute small workloads data processing can actually increase the compute time. Distributed computing will always require a certain amount of resources to perform scheduling and plan out the processing. For truly large data problems, distributed computing will rise above, even with the requirement of extra resources. Though, switching to a distributed computing solution requires code to be writen such that it can be processed in paralell (Even if a distributed computing solution reduces compute time, it may require much more engineering time). The process of writing code such that it can be processed in paralell can be done in any language, but some languages like Scala can efficiently analyze and optimize your code during compile. Scala stands for Scaleable-Language. Scala was made as an attempt to solve inconveniencies and issues with Java. Scala runs on the JVM, and is inter-operable with Java. Scala is most notably a functional language, but it does support OOP. Scala requires much less overhead than Java, code can be written in much fewer lines, Scala automatically performs optimization for paralell processing during compile, and Scala supports most Java libraries. Often times in Data Science and Computer Science, I feel that a solution's simplicity is undervalued. I feel like often times it is a better approach to optimize for program simplicity / readability / ease of maintainence are far more important than optimizing for runtime, unless the runtime of a program is truly important (User / Client facing applications should value quick runtime and user experience).
