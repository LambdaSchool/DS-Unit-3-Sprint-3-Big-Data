```scala
// defining Scala functions

// single expression function
def destroyAllHumans(human_count: Int, destroy_humans: Int): Int = human_count  - destroy_humans


// multiple expression
def countFruits(apples: Int, oranges: Int, bananas: Int): Int = {
    val apples_oranges = apples + oranges
    val oranges_bananas = oranges + bananas
    val apples_bananas = apples + bananas
    val all_fruit = apples + oranges + bananas
    all_fruit
}
```

```scala
countFruits(5, 10, 20)
```

```scala
destroyAllHumans(100, 50)
```

```scala
// if args in anon functions are used only once, define like so
val addOne: Int => Int = _ + 1
val weirdSum: (Int, Int) => Int = (_ * 2 + _ * 3)

addOne(9000)
weirdSum(6, 3)
```

```scala
// recursive functions need explicit return types, note: Unit == Void
def showNumbersInRange(a: Int, b: Int): Unit = {
    print(a)
    if (a < b)
        showNumbersInRange(a + 1, b)
}
showNumbersInRange(1, 14)
```
