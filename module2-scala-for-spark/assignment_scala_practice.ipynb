```scala
// def functionName(args...): ReturnType = { body... }
// In Scala, the last expression in the function block is the return value.
def sumOfSquares(x: Int, y: Int): Int = {
    val x2 = x*x
    val y2 = y*y
    x2+y2
}
```

```scala
//The {} can be omitted if the function body is a single expression
def sumOfSquaresShhort(x: Int, y: Int): Int = x*x + y*y
```

```scala
//syntax for calling functions is familiar
sumOfSquares(3,4)
```

```scala
//you can use parameters names to specify them in different order
def subtract(x: Int, y: Int): Int = x - y
subtract(10,3)
subtract(y=10,x=3)
```

```scala
// In most cases (with recursive functions the most notable exception), 
// function return type can be omitted
def sq(x: Int) = x*x
```

```scala
// functions can have default parameters
def addWithDefault(x: Int, y: Int = 5) = x + y
addWithDefault(1,2)
addWithDefault(1)
```

```scala
//anonymous function
(x: Int) => x*x
//in this function input (before the =>) and return (after =>) are specified
val sq: Int => Int = x => x*x
sq(10)
```

```scala
// if an argument is used onl once, there is a shorter way to define them:
val addOne: Int => Int = _ +1
val weirdSum: (Int, Int) => Int = (_*2 + _*3)
addOne(5)
weirdSum(2,4)
```
