## 2. Functions

```scala
def sumOfSquares(x: Int, y: Int): Int = {
    val x2 = x * x
    val y2 = y * y
    x2 + y2
}
```

```scala
def sumOfSquaresShort(x: Int, y: Int): Int = x * x + y * y
```

```scala
sumOfSquares(3, 4)
```

```scala
def subtract(x: Int, y: Int): Int = x - y
```

```scala
subtract(10, 3)
```

```scala
subtract(y=10, x=3)
```

```scala
def sq(x: Int) = x * x  // compiler can guess return type is Int
```

```scala
def addWithDefault(x: Int, y: Int = 5) = x + y
addWithDefault(1, 2)
addWithDefault(1)
```

```scala
(x: Int) => x * x   // anonymous function
```

```scala
val sq: Int => Int = x => x * x // Input types of anonymous functions can be left unspecified, if they can be inferred.
```

```scala
sq(10)
```

```scala
val addOne: Int => Int = _ + 1
val weirdSum: (Int, Int) => Int = (_ * 2 + _ + 3)
```

```scala
addOne(5)
weirdSum(2, 4)
```

## 3. Flow Control

```scala
(5 to 1 by -1) foreach println
```

```scala
def showNumbersInRange(a: Int, b: Int): Unit = {
    print(a)
    if (a < b)
        showNumbersInRange(a + 1, b)
}
showNumbersInRange(1, 14)
```

```scala
if (x == 1) println("yeah")
if (x == 10) print("yeah")
if (x == 11) println("yeah")
if (x == 11) println("yeah") else println("nay")
```

## Stretch Goals

```scala
val myNumbers = List(1, 2, 3)
```

```scala
import scala.math.pow
val cube: Int => Int = pow(_, 3).toInt
cube(2)
cube(3)
```

```scala
myNumbers map cube
```

```scala
myNumbers map (x => x * x * x)
```

```scala
def factorial(n: Int): Int = {
    n match {
        case 0 => 1
        case _ => n * factorial(n - 1)
    }
}
```

```scala
myNumbers foreach (x => println(factorial(x)))
```
