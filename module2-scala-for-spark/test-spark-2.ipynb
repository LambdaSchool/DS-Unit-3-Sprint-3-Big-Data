```scala
def sumOfSquares(x: Int, y: Int): Int = {
  val x2 = x * x
  val y2 = y * y
  x2 + y2
}
```

```scala
def sumOfSquaresShort(x: Int, y: Int): Int = x * x + y * y
```

```scala
sumOfSquares(5, 5)
```

```scala
sumOfSquaresShort(4, 5)
```
