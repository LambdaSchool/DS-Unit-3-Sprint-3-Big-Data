## 2. Functions 

```scala
// functions are defined like so
//
// def functionName(args...): ReturnType = {body....}
//
//if you come from more tradional languages, notice the omission of the 
//return keyword. In Scala, the last expression in the function block is the 
// return value.

def SumOfSquares(x :Int, y: Int): Int ={
    val x2 = x * x
    val y2 = y * y
    x2 + y2
}
```

```scala
//the {} can be ommited if the function body is a single expression:
def SumOfSquares(x: Int, y: Int): Int = x*x + y*y
```

```scala
//Syntax for calling functions is familiar:
SumOfSquares(3,4) // =>25
```
